```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
from matplotlib.legend_handler import HandlerLine2D
import matplotlib.lines as mlines
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv("data/city_data.csv")
ride_data = pd.read_csv("data/ride_data.csv")

city_data.head()
```

```python
# Tidy up the city data
city_type = city_data.drop_duplicates('city')
```

```python
ride_data.head()
```

```python
# Combine the data into a single dataset
city_ride_merger = ride_data.merge(city_data, on='city')

# Display the data table for preview
city_ride_merger.head()
```

```python
group_city = city_ride_merger.groupby(['city'])
```

## Bubble Plot of Ride Sharing Data

```python
#Average Fare ($) Per City
avg_fare = group_city['fare'].mean()

#Total Number of Rides Per City (rpc... like mpg but with rides per city)
rpc = group_city['ride_id'].count()

#Total Number of Drivers Per City
dpc = group_city['driver_count'].mean()

#City Type (Urban, Suburban, Rural)
#Realized I should clean up city_data
city_type_forreal = city_type.set_index('city')['type']
```

```python
# So fresh and so clean dataframe 
city_ride_forreal = pd.DataFrame({
    "Average Fare": avg_fare,
    "Rides per City": rpc,
    "Drivers per City": dpc,
    "City Types": city_type_forreal})

city_ride_forreal.head()
```

```python
# Shout out to those city types! Where they at?
rural = city_ride_forreal.loc[city_ride_forreal['City Types']=='Rural']
urban = city_ride_forreal.loc[city_ride_forreal['City Types']=='Urban']
suburban = city_ride_forreal.loc[city_ride_forreal['City Types']=='Suburban']
```

```python
# Build the scatter plots for each city types
plt.figure(figsize=(10,7))

#You must stick to the Pyber color scheme... WHY THOUGH?! Does color family count? 
r_plot = plt.scatter(x=rural['Rides per City'],
                        y=rural['Average Fare'],
                        s=rural['Drivers per City']*20,
                        color = 'lightcoral',edgecolor = 'k',
                        label = 'Rural', alpha = .6, linewidth= .75)

s_plot = plt.scatter(x=suburban['Rides per City'],
                           y=suburban['Average Fare'],
                           s=suburban['Drivers per City']*20,
                           color = 'gold',edgecolor = 'k',
                           label = "Suburban", alpha = .6,linewidth= 1)

u_plot = plt.scatter(x=urban['Rides per City'],
                        y=urban['Average Fare'],
                        s=urban['Drivers per City']*20,
                        color = 'lightskyblue',edgecolor = 'k',
                        label = "Urban", alpha = .6,linewidth= .75)


# Incorporate the other graph properties

plt.title('Pyber Ride Sharing Data (2016)', fontsize=20)
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.xlim(0, 41)

# Create a legend

legend = plt.legend(loc = "best", fontsize= 10, numpoints =1, 
                    scatterpoints=1, title= 'City Types', labelspacing = .5)
# make bubbles the same size
legend.legendHandles[0]._sizes = [50]
legend.legendHandles[1]._sizes = [50]
legend.legendHandles[2]._sizes = [50]

#reference regarding cicle size
plt.text(43,35, "Note: Circle size correlates with driver count per city.")
plt.grid(b=None, which='major', axis='both')

plt.show()

# Save Figure
plt.savefig('PyberBubble.png')
```

## Total Fares by City Type

```python

# % of Total Fares by City Type
total_fares = city_ride_merger.fare.sum()

# % of Total Rides by City Type
total_rides = city_ride_merger.ride_id.count()

# % of Total Drivers by City Type
total_drivers = dpc.sum()

print(total_fares, total_rides, total_drivers)
```

```python
# Calculate Fare Percents

r_fare = city_ride_merger.fare[city_ride_merger.type == 'Rural'].sum()
r_fare = round(r_fare/total_fares,4)

s_fare = city_ride_merger.fare[city_ride_merger.type == 'Suburban'].sum()
s_fare = round(s_fare/total_fares,4)

u_fare = city_ride_merger.fare[city_ride_merger.type == 'Urban'].sum()
u_fare = round(u_fare/total_fares,4)

#print(r_fare, s_fare, u_fare)

# Build Pie Chart
labels = ['Rural', 'Suburban', 'Urban']
sizes = [r_fare, s_fare, u_fare]
explode = (0, 0, .2)
colors = ["lightcoral", "gold", "lightskyblue"]

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title('% of Total Fares by City Type')
plt.axis('equal')

# Save Figure
plt.savefig("%Fares.png")
```

## Total Rides by City Type

```python
# Calculate Ride Percents
r_ride = city_ride_merger.ride_id[city_ride_merger.type == 'Rural'].count()
r_ride = round(r_ride/total_rides,4)

s_ride = city_ride_merger.ride_id[city_ride_merger.type == 'Suburban'].count()
s_ride = round(s_ride/total_rides,4)

u_ride = city_ride_merger.ride_id[city_ride_merger.type == 'Urban'].count()
u_ride = round(u_ride/total_rides,4)

#print(r_ride, s_ride, u_ride)

# Build Pie Chart
labels = ['Rural', 'Suburban', 'Urban']
sizes = [r_ride, s_ride, u_ride]
explode = (0, 0, .2)
colors = ["lightcoral", "gold", "lightskyblue"]

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title('% of Total Rides by City Type')
plt.axis('equal')

# Save Figure
plt.savefig("%Rides.png")
```

## Total Drivers by City Type

```python
# Calculate Driver Percents

r_drivers = city_ride_forreal['Drivers per City'][city_ride_forreal['City Types'] == 'Rural'].sum()
r_drivers = round(r_drivers/total_drivers,4)

s_drivers = city_ride_forreal['Drivers per City'][city_ride_forreal['City Types'] == 'Suburban'].sum()
s_drivers = round(s_drivers/total_drivers,4)

u_drivers = city_ride_forreal['Drivers per City'][city_ride_forreal['City Types'] == 'Urban'].sum()
u_drivers = round(u_drivers/total_drivers,4)

#print(r_drivers, s_drivers, u_drivers)

# Build Pie Chart

labels = ['Rural', 'Suburban', 'Urban']
sizes = [r_drivers, s_drivers, u_drivers]
explode = (0, 0, .2)
colors = ["lightcoral", "gold", "lightskyblue"]

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title('% of Total Drivers by City Type')
plt.axis('equal')

# Save Figure
plt.savefig("%Drivers.png")
```
